```python
%reload_ext autoreload
%autoreload 2

from stitching import *
from stitching import refineMatches_
import matplotlib.pyplot as plt
```

```python
# load images
images = loadImages(path='./images/images_01/')
```

```python
points_lst = getFeaturePoints(images, winSize=5, threshold=0.8, workers=10)
```

```python
descriptors_lst = getFeatureDesriptors(points_lst, images, winSize=16, num_bins=8, workers=10)
```

```python
matches_lst = match2Images(points_lst, descriptors_lst, threshold=4000, workers=10)
```

```python
H_lst = refineMatches(matches_lst, iteration=1000, threshold=0.8, workers=10)
```

```python
H_lst2 = H_lst.copy()
```

```python
H_lst2 = np.array(H_lst2)
```

```python
src_image, trg_image = images[0], images[1]
```

```python
f, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(src_image, cmap='gray')
ax[1].imshow(trg_image, cmap='gray')
```

```python
H = H_lst2[0].copy()
f1 = np.dot(H, np.array([0,0,1]))
f1 = f1/f1[-1]

H[0][-1] += abs(f1[0])
H[1][-1] += abs(f1[1])

offsety = abs(int(f1[1]))
offsetx = abs(int(f1[0]))

dsize = (trg_image.shape[0]+offsetx+100, trg_image.shape[1] + offsety)

dst = cv2.warpPerspective(src_image.T, H, 
                          dsize=dsize).T

dst[offsetx:offsetx+trg_image.shape[0],offsety:offsety+trg_image.shape[1]] = trg_image
plt.imshow(dst, cmap='gray')
```
